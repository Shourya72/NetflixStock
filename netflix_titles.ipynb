```python
import pandas as pd
```

```python
imdb = pd.read_csv('imdb_titles.tsv',sep='\t')
netflix = pd.read_csv('netflix_titles.csv')
ratings = pd.read_csv('ratings.tsv',sep='\t')
```

```python
imdb
```

```python
imdb.drop(['tconst', 'titleType', 'originalTitle', 'isAdult', 'endYear', 'runtimeMinutes', 'genres'], axis=1)
```

```python
netflix
```

```python
netflix = netflix.drop(['show_id', 'type', 'director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description'], axis=1)
```

```python
ratings
```

```python
imdb_ratings = pd.merge(imdb, ratings, on='tconst', how='inner')
```

```python
imdb_ratings = imdb_ratings.drop(['tconst', 'titleType', 'originalTitle', 'isAdult', 'endYear', 'runtimeMinutes', 'genres'], axis=1)
```

```python
imdb_ratings['primaryTitle_clean'] = imdb_ratings['primaryTitle'].str.lower().str.strip()
netflix['title_clean'] = netflix['title'].str.lower().str.strip()
```

```python
imdb_no_dup = imdb_ratings.sort_values('numVotes', ascending=False).drop_duplicates(subset='primaryTitle_clean') # Remove duplicates, keep instance with greatest number of votes.
```

```python
netflix_titles = pd.merge(imdb_no_dup, netflix, left_on='primaryTitle_clean', right_on='title_clean', how='inner')
```

```python
netflix_titles['date_added'] = netflix_titles['date_added'].str.strip()
netflix_titles['date_added'] = pd.to_datetime(netflix_titles['date_added'], errors='coerce')
```

```python
netflix_titles
```

```python
import matplotlib.pyplot as plt

df = netflix_titles.dropna(subset=['averageRating', 'date_added'])

df['year_month'] = df['date_added'].dt.to_period('M')

ratings_over_time = df.groupby('year_month')['averageRating'].mean().reset_index()
ratings_over_time['year_month'] = ratings_over_time['year_month'].dt.to_timestamp()

plt.figure(figsize=(12, 6))
plt.plot(ratings_over_time['year_month'], ratings_over_time['averageRating'], marker='o')
plt.title('Average IMDb Rating of Netflix Titles Over Time')
plt.xlabel('Date Added (Year-Month)')
plt.ylabel('Average IMDb Rating')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['year_month'] = df['date_added'].dt.to_period('M')

titles_per_month = df.groupby('year_month').size().reset_index(name='count')
titles_per_month['year_month'] = titles_per_month['year_month'].dt.to_timestamp()

plt.figure(figsize=(12, 6))
plt.bar(titles_per_month['year_month'], titles_per_month['count'], width=20)
plt.title('Number of Netflix Titles Added Per Month')
plt.xlabel('Date Added (Year-Month)')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
